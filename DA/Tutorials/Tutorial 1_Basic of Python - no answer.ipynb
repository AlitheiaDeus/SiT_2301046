### No 2: List & Indexing
Lists allow us to hold multiple values in a single container. A list can contain values of any data type.

```python
#create lists
list_1 = [10, 20, 30, 40, 50, 60]
list_2 = ['A', 'B', 'C', 'D', 'E', 'F']
```

```python
#we can use indexing & slicing (just like with strings!) to access individual values in the list
print(list_1[3])
```

```python
# Assign and access individual values 

new_list = ['A', 'B', 'C', 'D', 'E', 'F']
new_list[0] = 'Z'
print(new_list)
```

```python
#list length - len()
new_list = ['A', 'B', 'C', 'D', 'E', 'F']
print(len(new_list))
```

```python
#append elements to a list
new_list = ['A', 'B', 'C', 'D', 'E', 'F']

new_list.append('G')
print(new_list)
```

```python
#remove elements from a list
#new_list = ['A', 'B', 'C', 'D', 'E', 'F']

new_list.remove('A')
print(new_list)
```

```python
#Deleting by index - del,index start from ‘0’

new_list = ['A', 'B', 'C', 'D', 'E', 'F']

del new_list[1]
print(new_list)
```

### List - Sorting 
The sorted function sorts the values of a list in ascending order and returns a new list of sorted values.

```python
#create 2 list
new_list = [1, 2, 3, 8, 10, 9]
alphabets = ["A", "C", "B"]

#by default: ascending order
print(sorted(new_list)) 
print(sorted(alphabets))

#descending order
print(sorted(new_list, reverse=True)) 
print(sorted(alphabets, reverse=True))
```

### No 3:IF-ELIF-ELSE


```python
#IF-ELIF-ELSE
temperature = 100 
if temperature < 150: 
    print("temperature is too low!") 
elif temperature > 200: 
    print("temperature is too high!") 
else: 
    print("good temperature control.")
```

```python
#IF-ELIF-ELSE Example
data = ['product A', 'product B', 'product C', 
 'product D', 'product E', 'product F',
 'product G', 'product H', 'product I']

for idx in range(len(data)):
    
    if idx < 3:
        quality = "good"
    elif idx <= 5:
       quality = "rework"
    else:
       quality = "scrap"
        
    print(quality, "-", data[idx])
```

### No 4: Functions
* Functions are like chunks of pre-written code that we can use again and again.<br>
* We can "call" functions to perform an action for us.<br>
* Python's built-in functions: <br>


#### Why we want to build functions: 
* Makes code reusable
* Provides structure for longer programmes
* Makes it easier for other programmers to understand your work<br>


#### Components of a function:
* Name
* Parameters _input_
* Returned value _output_



```python
# Create the y_1 function:
def y_1(u, a, c):
    return u * c + a

# Call the y_1 function:
result = y_1(12, 10, 5)
print('The result is:', result)
```

```python
#Write a Function smaller_of_two() 
def smaller_of_two(num_1, num_2):
    if num_1 < num_2:
        result = num_1
    else:
        result = num_2
    return result

#call a funciton smaller_of_two() 
test_result = smaller_of_two(num_1=150, num_2=10) # keyword arguments
print("My test result is:", test_result)
test_result_2 = smaller_of_two(99, 200) # default order, i.e. num_1=99, num_2=11
print("My test result 2 is:", test_result_2)
```

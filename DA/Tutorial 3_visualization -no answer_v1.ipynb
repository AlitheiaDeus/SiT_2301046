```python
# import library
import numpy as np
import matplotlib.pyplot as plt
#prepare data x
x = np.linspace(0, 10, 100)

# prepare figure
fig = plt.figure()

#generate 2 plots
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.title("Line plot")
plt.xlabel('x-value')
plt.ylabel('y-value')
# save figure as .png in the current working directory
fig.savefig('my_figure.png')
```

```python
#subplot

import numpy as np
#use grid style
plt.style.use('seaborn-v0_8-whitegrid')

fig = plt.figure()
#Define how many subplots
fig, ax = plt.subplots(2)

# ax will be an array of 'n' Axes objects, means 'n' suplot
ax[0].plot(x, np.sin(x))
ax[0].title.set_text('Sin(x)')
ax[0].set(ylabel='Sin(x)')

ax[1].plot(x, np.cos(x));
ax[1].title.set_text('Cos(x)')
ax[1].set(xlabel='x', ylabel='Cos(x)')
#save file
fig.savefig('my_figure.png')
```

### Designing Your Plots
To make the visualizations more informative and effective: 
* What is the story you want to tell?
* Is the visualization easy to understand?
* Is the visualization/chart appropriate based on the given dataset? 

```python
#Control the line colors  draw 6 different color lines:

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supporte
```

```python
#set line style, color will generate randomly

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
#Add axes limit , title and Labeling for the plots

#set lable and style
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
#set x and y limit
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
#add the title and label
plt.title("Sine and Cosine Curve")
plt.xlabel("x")
plt.legend();
```

```python
#Scatter plot: scatter plots are best used for showing relationships between 2 variables.

import matplotlib.pyplot as plt
# generate 3 data arraies x, y, z
x = [1, 2, 3, 4, 5]
y = [1, 4, 6, 8, 10]
z = [1, 8, 27, 64, 125]
```

```python
# scatter plot:
plt.title("Scatter Plot")
plt.scatter(x,z, color='blue')
plt.xlabel('x-value')
plt.ylabel('z-value')
```

Scatter Plots - Color, Markers, Alphas, Size, When designing scatterplots, sometimes you want certain points to be more distinct than the others.

* Color c=
* Markers - Design of each point:  marker=
* Alphas - Darkness of each point:  alpha=
* Size - Size of each point:  s=(alpha <=1)

```python
plt.scatter(x , y , c= 'red', marker = '*', s = 50, alpha=1)
plt.title("Scatter Plot")
plt.xlabel('x')
plt.ylabel('y')
```

```python
#Bar charts and Column chart are used to view differences for categories or for time-series data.
#plot with Column chart
import matplotlib.pyplot as plt
# Generate data array x and y
x = [1, 2, 3, 4, 5]
y = [1, 4, 6, 5, 10]
#set the plot
plt.title("Column chart")
plt.bar(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#plot with bar chart
import matplotlib.pyplot as plt

#set the plot
plt.title("Bar chart")
plt.barh(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#plot histogram

import numpy as np
import matplotlib.pyplot as plt
#using white background
plt.style.use('seaborn-v0_8-white')
fig = plt.figure()
#generate 1000 random data
data = np.random.randn(1000)

plt.title("Histogram")
#Histogram plot using default setting, 10 bins.
plt.hist(data)
fig.savefig('my_figure.png')
```

```python
#Customize the histogram plot with parameters settings
plt.title("Histogram")
plt.hist(data, bins=20, alpha=0.8,
                histtype='stepfilled', color='blue', edgecolor='Red')
#seperate into 20 bins and generate y and x arrays
```

```python
# Sometimes it is better to have plots side-by-side to compare 2 different data sets.
#generate 3 data arrays
x = [1, 2, 3, 4, 5]
y= [1,4,6,8,10]
z = [1, 8, 27, 64, 125]

# set fig size
fig = plt.figure(figsize=(12, 6))

#1st row, 1st column, 1st chart
ax1 = fig.add_subplot(1, 2, 1) # 1x2 grid, 1st subplot
ax1.scatter(x, y)
plt.title("LEFT CHART x-y")

#1st row, 2nd column, 2nd chart
ax2 = fig.add_subplot(1, 2, 2)  # 1x2 grid, 2nd subplot
ax2.scatter(x, z, color='r')
plt.title("RIGHT CHART x-z")
```

```python
#subplot panda series with panda function.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 1)

#gererate a data series
data = pd.Series(np.random.rand(11), index=list('abcdefghijk'))

#print(data)

#plot bar and horizontal bar chart

data.plot.bar(ax=axes[0], color='r', alpha=0.7)
data.plot.barh(ax=axes[1], color='g', alpha=0.7)
```

```python
#subplot panda series using matplotlib.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#define figure size
fig = plt.figure(figsize=(12, 6))
# Top left chart
ax = fig.add_subplot(2, 2, 1) # row x col grid, nth plot index
ax.barh(data.index, data.values) 
plt.title("Bar chart")

# Top right chart
ax2 = fig.add_subplot(2, 2, 2)
ax2.bar(data.index, data.values, color='r')
plt.title("Column chart")

# Bottom chart
#ax3 = fig.add_subplot(2, 2, 3)
ax3 = fig.add_subplot(2, 1, 2)
ax3.plot(data.index, data.values, color='g')
plt.title("Line chart")
```

```python
#Plotting Pandas Dataframe
#Bar plots group the values in each row together in a group in bars, side by side, for each value. 
#create a datafram
df = pd.DataFrame(np.random.rand(6, 4),
index=['Customer 1', 'Customer 2', 'Customer 3', 'Customer 4', 'Customer 5', 'Customer 6'],
columns=pd.Index(['machine A', 'machine B', 'machine C', 'machine D'], name='Customers'))
print(df)
#plot cluster column chart
df.plot.bar().legend(bbox_to_anchor=(1.3, 1))
```

```python
#plot the stacked bar chart
df.plot.barh(stacked=True, alpha=0.8).legend(bbox_to_anchor=(1.2, 1))
```

```python
#Plot data with Seaborn - Scatter plot and fit a linear regression model.

# Load library
import seaborn as sns
import pandas as pd

# Create path
path = 'data_scatter_plot.xls'

# Load data
dataframe = pd.read_excel(path)

#scatter plot with linear regression model fit
sns.regplot(x='Rejected Parts', y='Total Production',data= dataframe)
```

```python
# Plot with Seaborn -Pairplot
# Seaborn has a convenient pairplot function, which supports placing histograms or density estimates of each variable along the diagonal.

# Load library
import seaborn as sns
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Create path
path = 'lesson3_data.csv'
#list of column title in the excel 
#print (dataframe.columns.ravel())

# Load data
dataframe = pd.read_csv(path)
print(dataframe)

#pari plot
sns.pairplot(dataframe)
```

```python
#calculate correlation and validate linear relationship
corr = dataframe.select_dtypes(include=[np.number]).corr(method='pearson')
corr = round(corr, 3)
corr.style.background_gradient(cmap='coolwarm')
```

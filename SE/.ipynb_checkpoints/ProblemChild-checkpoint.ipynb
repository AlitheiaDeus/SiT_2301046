```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('RainfallMonthlyNumberofRainDays.csv')

display(df)
```

```python
df[['yyyy', 'mm']] = df['month'].str.split('-', expand=True)

del df['month']

display(df)
```

```python
ys = df['yyyy'].unique()
ms = df['mm'].unique()

df_des = {}

for y in ys:
    df_des[y] = df[df['yyyy'] == y].describe(include = 'all')

for y in ys:
    print('V - ' + y + ' - V')
    display (df_des[y])
```

```python
plt.figure(figsize = (12, 9))

for y in ys:
    plt.plot(df[df['yyyy'] == y]['mm'], df[df['yyyy'] == y]['no_of_rainy_days'], label = y)

plt.title('Rain')
plt.legend(bbox_to_anchor = (1.04, 0.5), loc = "center left")
plt.show()
```

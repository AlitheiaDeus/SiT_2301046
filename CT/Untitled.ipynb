```python
import matplotlib.pyplot as plt
from scipy.signal import tf2zpk
import numpy as np
import control as ct
```

```python
w_n = 4
zeta = 0.5

# Define the transfer function (assuming a SISO system)
output_f = [1]
input_f = [1, 2, 0]

# Extract poles and zeros
z, p, k = tf2zpk(output_f, input_f)
print('-:-')
print(z)
print(p)
print(k)

sys = ct.tf(output_f, input_f)
print('-:-')
print(sys)

print('-:-')
print(sys.poles)
```

```python
# Plot poles and zeros
plt.figure(figsize=(8, 8))
plt.plot(np.real(z), np.imag(z), 'o', markersize=8, label='Zeros')
plt.plot(np.real(p), np.imag(p), 'x', markersize=8, label='Poles')

# Add labels and title
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Poles and Zeros of the Transfer Function')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Convert to polar coordinates (magnitude and phase)
z_mag, z_phase = np.abs(z), np.angle(z)
p_mag, p_phase = np.abs(p), np.angle(p)

# Generate angles for the polar grid (0 to 2*pi)
theta = np.linspace(0, 2*np.pi, 200)

# Create the polar grid manually
plt.figure(figsize=(15, 15))
plt.polar(theta, np.ones_like(theta), color='lightgray', linestyle='--', linewidth=0.5)  # Circles

# Plot poles and zeros
plt.polar(z_phase, z_mag, 'o', markersize=15, label='Zeros')
plt.polar(p_phase, p_mag, 'x', markersize=15, label='Poles')

# Add labels and title
plt.xlabel('Angle (rad)')
plt.ylabel('Magnitude')
plt.title('Poles and Zeros on Polar Grid (Bode Plot)')

# Legend
plt.legend()
plt.show()
```

```python
import math
from math import pi as PI
import random

import numpy as np
import sympy as sp
from scipy import signal

import matplotlib as mpl
from matplotlib import pyplot as plt

import control as ct
from control import tf, bode_plot, nyquist_plot, pzmap

import sympy as sp  # for symbolic representation of variables
```

```python
Dt = 0.16743
Jp = 23188500.45 / 1000 * (0.001 ** 2)
Jy = 23810415.95 / 1000 * (0.001 ** 2)
ksp = 0.00744
Dp = 0.00199
Dy = 0.00192
kpp = 0.00321
kpy = 0.00137
kyy = 0.00610
kyp = -0.00319
v = (0.05 / 12) * 11 + 0.9955

MyVars = {
    'Dt': Dt,
    'Jp': Jp,
    'Jy': Jy,
    'ksp': ksp,
    'Dp': Dp,
    'Dy': Dy,
    'kpp': kpp,
    'kpy': kpy,
    'kyy': kyy,
    'kyp': kyp,
    'v': v
}
display(MyVars)

# G1 = 
# G2 = 
# G3 = 
# G4 = 
```

```python
# Define transfer function
num1 = [kpp*Dt]  # numerator coefficients
den1 = [Jp, Dp, v*ksp]  # denominator coefficients
sys1 = tf(num1, den1)

# Plot pole/zero diagram
pz1 = pzmap(sys1)

response1 = ct.forced_response(sys1, T = np.linspace(0,30,100), U = [2 for _ in range(100)])

ReOp1 = []
for i in response1.outputs:
    ReOp1.append(i * 180 / PI)

fig = plt.figure(figsize = (8, 5))
plt.plot(response1.time, ReOp1)
plt.grid()
plt.show()

fig = plt.figure(figsize = (8, 5))
gl = bode_plot(sys1)
plt.show()

fig = plt.figure(figsize = (8, 5))
gl = nyquist_plot(sys1)
plt.show()

# Define transfer function
num2 = [kyp*Dt]  # numerator coefficients
den2 = [Jy, Dy, 0]  # denominator coefficients
sys2 = tf(num2, den2)

# Plot pole/zero diagram
pz2 = pzmap(sys2)

response2 = ct.forced_response(sys2, T = np.linspace(0,30,100), U = [2 for _ in range(100)])

ReOp2 = []
for i in response2.outputs:
    ReOp2.append(i * 180 / PI)

fig = plt.figure(figsize = (8, 5))
plt.plot(response2.time, ReOp2)
plt.grid()
plt.show()

fig = plt.figure(figsize = (8, 5))
gl = bode_plot(sys2)
plt.show()

fig = plt.figure(figsize = (8, 5))
gl = nyquist_plot(sys2)
plt.show()
```
